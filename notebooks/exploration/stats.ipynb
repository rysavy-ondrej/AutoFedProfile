# Basic statistic on prepared datasets

```python
from pathlib import Path

import pandas as pd

data_dir = Path("../data")
label_files = sorted(data_dir.glob("*_labels.parquet"))
categories = ["system", "application", "malware", "unknown"]

rows = []
for label_file in label_files:
    labels = pd.read_parquet(label_file)["category"]
    counts = labels.value_counts()
    row_name = label_file.name.replace("_labels", "")
    rows.append({"parquet_file": row_name, **{category: int(counts.get(category, 0)) for category in categories}})

stats_df = (
    pd.DataFrame(rows)
    .set_index("parquet_file")
    .loc[:, categories]
    .sort_index()
)

stats_df
```
